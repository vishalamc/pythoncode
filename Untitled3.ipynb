```python
import numpy as np
```

```python
list1=[]

ans=True
while ans:
    print("Scientific Calculator")
    print("************************")
    print("1 2 3 sin()\n4 5 6 cos()\n7 8 9 tan()\n0 x result")
    print("************************")
    ans=input("Enter your choice\n")
    if ans=="1": 
        list1.append(1) 
    elif ans=="2":
        list1.append(2) 
    elif ans=="3":
        list1.append(3)
    elif ans=="4":
        list1.append(4) 
    elif ans=="5":
        list1.append(5) 
    elif ans=="6":
        list1.append(6) 
    elif ans=="7":
        list1.append(7)
    elif ans=="8":
        list1.append(8) 
    elif ans=="9":
        list1.append(9) 
    elif ans=="0":
        list1.append(0)
    elif ans=="sin":
        a=np.array(list1)
        b=np.sin(a)
    elif ans=="x":
        break
    elif ans=="result":
        print(b) 
        break
    elif ans !="":
        print("\nInvalid choice") 
```
