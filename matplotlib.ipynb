```python
import matplotlib.pyplot as plt
x=['a','b','c','d']
y=[1,4,9,16]
z=[2,3,4,5]
#plot 1
plt.subplot(2,2,1)
plt.title("First Plot")
plt.plot(x,y)
#plot 2
plt.subplot(2,2,2)
plt.title("Second plot")
plt.plot(x,z)

plt.suptitle("My Sub Plots")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=['a','b','c','d']
y=[11,14,19,16]
z=[2,3,4,5]
#plot 1
plt.subplot(1,2,1)# 1 row 2 cols index=1
plt.title("First Plot")
plt.plot(x,y)
#plot 2
plt.subplot(1,2,2)# 1 row 2 cols index=2
plt.title("Second Plot")
plt.plot(x,z)

plt.show()
```

```python
import matplotlib.pyplot as plt
x=['a','b','c','d']
y=[11,14,19,16]
z=[2,3,4,5]
#plot 1
plt.subplot(2,2,1)# 2 row 2 cols index=1
plt.title("First Plot")
plt.plot(x,y)
#plot 2

plt.subplot(2,2,2)# 2 row 2 cols index=2
plt.title("Second Plot")
plt.plot(x,z)

#plot 3
plt.subplot(2,2,3)# 2 row 2 cols index=3
plt.title("third Plot")
plt.plot(y,z)

#plot 4
plt.subplot(2,2,4)# 2 row 2 cols index=4
plt.title("fourth Plot")
plt.plot(y,x)

plt.show()
```

```python
# box  plot example
import matplotlib.pyplot as plt
import numpy as np
x=[10,20,30,40,50,60]
print(np.min(x))
print(np.max(x))
print(np.quantile(x,[0.25,0.50,0.75]))
plt.boxplot(x)
plt.title('Student details')
plt.show()
```
